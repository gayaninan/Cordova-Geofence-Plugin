```python
df_refs = df_refs.rename(columns = {'ref_index':'corpus_index'})

w_refs = w_refs.rename({'ref_index': 'ref_index',
                'aa': 'aa_wer', 'am':'am_wer', 'ai' : 'ai_wer', 'ag' : 'ag_wer', 
                'ma' : 'ma_wer', 'mm': 'mm_wer','mi':'mi_wer', 'mg':'mg_wer',
                'ia':'ia_wer','im':'im_wer', 'ii':'ii_wer', 'ig':'ig_wer',
                'ga':'ga_wer', 'gm':'gm_wer', 'gi':'gi_wer', 'gg':'gg_wer', 
                'label':'label_wer'}, axis=1)

r1_refs = r1_refs.rename({'ref_index': 'ref_index',
                'aa': 'aa_rouge1', 'am':'am_rouge1', 'ai' : 'ai_rouge1', 'ag' : 'ag_rouge1', 
                'ma' : 'ma_rouge1', 'mm': 'mm_rouge1','mi':'mi_rouge1', 'mg':'mg_rouge1', 
                'ia':'ia_rouge1','im':'im_rouge1', 'ii':'ii_rouge1', 'ig':'ig_rouge1',
                'ga':'ga_rouge1', 'gm':'gm_rouge1', 'gi':'gi_rouge1', 'gg':'gg_rouge1', 
                'label':'label_rouge1'}, axis=1)

r2_refs = r2_refs.rename({'ref_index': 'ref_index',
                'aa': 'aa_rouge2', 'am':'am_rouge2', 'ai' : 'ai_rouge2', 'ag' : 'ag_rouge2', 
                'ma' : 'ma_rouge2', 'mm': 'mm_rouge2','mi':'mi_rouge2', 'mg':'mg_rouge2', 
                'ia':'ia_rouge2','im':'im_rouge2', 'ii':'ii_rouge2', 'ig':'ig_rouge2',
                'ga':'ga_rouge2', 'gm':'gm_rouge2', 'gi':'gi_rouge2', 'gg':'gg_rouge2', 
                'label':'label_rouge2'}, axis=1)

```

```python
# def most_common(lst):
#     return max(set(lst), key=lst.count)
```

```python
from collections import Counter

def get_oc(lst):
  print(lst)
  c = Counter(lst)

  d = c.most_common(1)
  print(d[0][0])
  return d[0][0]

def get_oc_stat(lst):
  print(lst)
  c = Counter(lst)

  # d = c.most_common(1)
  # print(d[0][0])


  if len(c) == 3:
    return 3
  elif len(c) == 2: 
    return 2
  elif len(c) == 1:
    return 1
```

```python
merged_evals = w_refs.merge(r1_refs).merge(r2_refs)
merged_evals
```

```python
final_merge = df_refs.join(merged_evals)
final_merge
```

```python
# final_merge.to_csv('/content/drive/MyDrive/updated-ec/test/merged_class_updated.txt', sep='\t', index=False )
```

```python
#test
final_merge['common_label'] = final_merge.apply(lambda row: get_oc([row.label_wer, row.label_rouge1, row.label_rouge2]), axis=1)
final_merge['no_of_labels'] = final_merge.apply(lambda row: get_oc_stat([row.label_wer, row.label_rouge1, row.label_rouge2]), axis=1)
```
